## 3. Gamma es previa conjugada de Poisson

Supongamos que tenemos un dato $x \in \mathbb{N}$ que tiene una verosimilitud de $\text{Poisson}(\theta)$:

$$
p(x | \theta) = \frac{\theta^x e^{-\theta}}{x!}.
$$

Ahora, si elegimos la distribución previa sobre el parámetro $\theta$ como una distribución $\text{Gamma}(a, b)$:

$$
p(\theta) = \frac{b^a}{\Gamma(a)} \theta^{a - 1} \exp\{-b \theta\}
$$

Encontrar completamente la densidad de probabilidad posterior

$$
p(\theta|x) \propto \underbrace{\left( \theta^x e^{-\theta} \right)}_{\text{Verosimilitud}} \cdot \underbrace{\left( \theta^{a-1} e^{-b\theta} \right)}_{\text{Previa}}
$$

Agrupar términos de $\theta$ 

Ahora simplemente sumamos los exponentes de las bases iguales:

Para $\theta$: Tenemos $x$ y $a-1$, lo que nos da $\theta^{a+x-1}$.

Para $e$: Tenemos $-\theta$ y $-b\theta$, lo que factorizado es $e^{-(b+1)\theta}$.

$$
p(\theta|x) \propto \theta^{a+x-1} e^{-(b+1)\theta}
$$

Nueva distribución

Al ver la forma anterior notamos que tiene exactamente la estructura de una distribución $\text{Gamma}(a', b')$.

Por lo tanto, definimos los nuevos parámetros actualizados:

Nuevo parámetro de forma ($a'$): $a + x$

Nuevo parámetro de tasa ($b'$): $b + 1$

Escribir la Densidad Posterior Completa

Para que deje de ser solo una "proporcionalidad" y sea una densidad de probabilidad válida, añadimos la constante de normalización de la Gamma con los nuevos parámetros:

$$
p(\theta|x) = \frac{(b+1)^{a+x}}{\Gamma(a+x)} \theta^{a+x-1} \exp\{-(b+1)\theta\}
$$

## 4. Fórmulas de actualización normal

Supongamos que tenemos un dato $x \sim \text{Normal}(\theta, \sigma^2)$, donde la varianza $\sigma^2$ es conocida. Es decir, la media $\theta$ es nuestro parámetro deconocido de interés. Si elegimos una previa normal

$$
\theta \sim \text{Normal}(\mu_0, \sigma_0^2),
$$

entonces la posterior también es normal $(\theta | x) \sim \text{Normal}(\mu_1, \sigma_1^2)$. Demostrar que

$$
\frac{\mu_1}{\sigma_1^2} = \frac{\mu_0}{\sigma_1^2} + \frac{x}{\sigma^2}, \qquad \frac{1}{\sigma_1^2} = \frac{1}{\sigma_0^2} + \frac{1}{\sigma^2},
$$

o equivalentemente

$$
\mu_1 = \frac{\mu_0 / \sigma_0^2 + x / \sigma^2}{1 / \sigma_0^2 + 1 / \sigma^2} = \frac{\sigma^2 \mu_0 + \sigma_0^2 x}{\sigma_0^2 + \sigma^2}, \qquad \sigma_1^2 = \frac{1}{1 / \sigma_0^2 + 1 / \sigma^2} = \frac{\sigma_0^2\sigma^2}{\sigma_0^2 + \sigma^2}
$$

Aplicar el Teorema de Bayes

Multiplicamos la verosimilitud por la previa y nos enfocamos únicamente en lo que está dentro de la exponencial (el exponente):
$$
p(\theta|x) \propto \exp\left\{ -\frac{1}{2} \left[ \frac{(x-\theta)^2}{\sigma^2} + \frac{(\theta-\mu_0)^2}{\sigma_0^2} \right] \right\}
$$

Expandir y agrupar términos de $\theta$ 

Para identificar la nueva media y varianza, expandimos los cuadrados dentro del corchete:
$$
\text{Exponente} = \frac{x^2 - 2x\theta + \theta^2}{\sigma^2} + \frac{\theta^2 - 2\theta\mu_0 + \mu_0^2}{\sigma_0^2}
$$

Agrupamos los términos según la potencia de $\theta$:

Términos con $\theta^2$: $\theta^2 \left( \frac{1}{\sigma^2} + \frac{1}{\sigma_0^2} \right)$

Términos con $\theta$: $-2\theta \left( \frac{x}{\sigma^2} + \frac{\mu_0}{\sigma_0^2} \right)$

Términos constantes: (Se ignoran).

Demostrar las fórmulas de actualización

Sabemos que una distribución Normal con media $\mu_1$ y varianza $\sigma_1^2$ tiene un exponente de la forma:
$$
\frac{(\theta - \mu_1)^2}{\sigma_1^2} = \frac{\theta^2}{\sigma_1^2} - \frac{2\theta\mu_1}{\sigma_1^2} + \text{const}
$$

Comparando término a término con lo que obtuvimos:

Para la varianza: El coeficiente de $\theta^2$ debe ser $1/\sigma_1^2$.
$$
\frac{1}{\sigma_1^2} = \frac{1}{\sigma_0^2} + \frac{1}{\sigma^2}
$$

Para la media: El coeficiente de $-2\theta$ debe ser $\mu_1/\sigma_1^2$.
$$
\frac{\mu_1}{\sigma_1^2} = \frac{\mu_0}{\sigma_0^2} + \frac{x}{\sigma^2}
$$

Si despejas $\mu_1$ y $\sigma_1^2$ de las ecuaciones anteriores, llegas exactamente a las fórmulas de tu imagen:
$$
\mu_1 = \frac{\frac{\mu_0}{\sigma_0^2} + \frac{x}{\sigma^2}}{\frac{1}{\sigma_0^2} + \frac{1}{\sigma^2}} = \frac{\sigma^2 \mu_0 + \sigma_0^2 x}{\sigma_0^2 + \sigma^2} \quad \text{y} \quad \sigma_1^2 = \frac{\sigma_0^2 \sigma^2}{\sigma_0^2 + \sigma^2}
$$
